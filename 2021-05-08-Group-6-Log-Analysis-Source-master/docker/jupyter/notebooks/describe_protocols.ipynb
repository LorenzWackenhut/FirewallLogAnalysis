```python
import pandas as pd
import matplotlib.pyplot as plt
import pyarrow.parquet as pq
```

```python
# Full parquet needs to be build
# df = pd.read_parquet('./data/df_v4.parquet', engine='pyarrow') 
df = pd.read_parquet('./data/df_v4_10000', engine='pyarrow') 
```

```python
# Plot count for each protocol
ax = df['protocoll_text'].value_counts().plot(kind='bar',figsize=(20,4), width=0.3)
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
```

# Documentation of protocols
Determine communication between client and server.

- tcp = Transmission Control Protocol. Reliable, ordered but heavyweight handling. Mostly used for HTTP, FTP, SMTP
- udp = User Datagram Protocol. Unreliable, not ordered but lightweight handling. Mostly used for video and audio traffic
- ff02::2 = ipv6 Multicast (Link-Local Scope Multicast Addresses)
- icmp = Internet Control Message Protocol. Typically used or diagnostic or control purposes, also error handling (Exploit https://resources.infosecinstitute.com/topic/icmp-attacks/)
- sctp = Stream Control Transmission Protocol. Improvement over TCP and UDP, more secure, better reliability
- gre = Generic Routing Encapsulation. Tunneling protocol. Connect one intranet to the other e.g.
- ipv6
- igmp = Internet Group Management Protocol. Spotify, Netflix send data streams to entire receiver groups
- pim = Protocol Independent Multicast. Provide one-to-many and many-to-many distribution of data over LAN, WAN or Internet
- nvp = Network Voice Protocol. Transport human speech.

```python
icmp_df = df[df['protocoll_text'] == 'icmp']
ax = icmp_df['source_ip'].value_counts()[:50].plot(kind='bar',figsize=(20,4), width=0.3)
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
```
